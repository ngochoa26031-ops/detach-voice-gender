# detach-voice-gender on Kaggle

Notebook nay chi la bootstrap nhe. Ban import 1 lan, ve sau chi can bam `Run`.

Moi lan chay, notebook se tu:

- clone repo `ngochoa26031-ops/detach-voice-gender`
- chay `run_kaggle.py` moi nhat tu GitHub

Truoc khi chay (chi can lam 1 lan):

1. `Add-ons -> Secrets -> them secret ten HF_TOKEN` (token HuggingFace, da accept license `pyannote/speaker-diarization-3.1` va `pyannote/segmentation-3.0`)
2. `Settings -> Accelerator -> GPU`
3. `Settings -> Internet -> On`
4. (Tuy chon) Neu muon backup output/input/resume len Google Drive: xem cell ben duoi, dien ten remote rclone cua ban vao, va them Kaggle Secret `RCLONE_CONF_B64` (xem `get_rclone_secret.bat` trong repo).
5. Bam `Run All` de mo app Gradio, upload video/audio + file .srt truc tiep tren web, bam chay, tai ket qua ve ngay tren web

```python
# (Tuy chon) Backup Google Drive qua rclone. Bo qua/xoa cell nay neu khong can.
# Doi "gdrive" thanh dung ten remote ban da tao bang `rclone config`,
# va phai co Kaggle Secret RCLONE_CONF_B64 (xem README.md muc Drive backup).
import os

RCLONE_REMOTE_NAME = "gdrive"  # <-- doi ten nay cho dung remote cua ban

os.environ["GENDERSFX_RCLONE_REMOTE"] = f"{RCLONE_REMOTE_NAME}:detach-voice-gender/output"
os.environ["GENDERSFX_RCLONE_INPUT_REMOTE"] = f"{RCLONE_REMOTE_NAME}:detach-voice-gender/input"
os.environ["GENDERSFX_RCLONE_RESUME_REMOTE"] = f"{RCLONE_REMOTE_NAME}:detach-voice-gender/resume"
```

```python
import subprocess, sys
from pathlib import Path

ROOT = Path('/kaggle/working')
LOADER_REPO = ROOT / 'detach_voice_gender_loader'

print('[*] Xoa loader cu neu co...', flush=True)
subprocess.run(['rm', '-rf', str(LOADER_REPO)], check=False)
print('[*] Dang clone loader tu GitHub...', flush=True)
subprocess.run(['git', 'clone', '--depth', '1', 'https://github.com/ngochoa26031-ops/detach-voice-gender.git', str(LOADER_REPO)], check=True, timeout=300)
print('[*] Dang chay bootstrap...', flush=True)
subprocess.run([sys.executable, '-u', str(LOADER_REPO / 'run_kaggle.py')], check=True)
```
